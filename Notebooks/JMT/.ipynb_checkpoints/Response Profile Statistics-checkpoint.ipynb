```python
#1 = Both
#2 = Stim_CR
#3 = Choice_CR
#4 = NCR

class_preearly = []
class_prelate = []
class_preexpert = []
class_postearly = []
class_postlate = []
class_postexpert = []

for idx,(session,clust) in tqdm(enumerate(zip(EnumSession,EnumClust))):
    sessionfile = ilep.loadSessionCached(act_directory,session)
    
    FRmod = sessionfile.responsiveness[clust]['all_trials'].FRmodulation
    RampingNorm = np.abs(sessionfile.responsiveness[clust]['go'].rampingModNorm)
    if not (np.isfinite(FRmod) and np.isfinite(RampingNorm)):
        continue
    
    stim_CR = (FRmod >= 3.5)
    choice_CR = (RampingNorm >= 0.5)
    
    this_class = np.nan
    if stim_CR and choice_CR:
        this_class = 1
    elif stim_CR and not choice_CR:
        this_class = 2
    elif not stim_CR and choice_CR:
        this_class = 3
    elif not stim_CR and not choice_CR:
        this_class = 4
        
        
    if EILphases[idx] == 'pre early':
        class_preearly.append(this_class)
    elif EILphases[idx] == 'pre late':
        class_prelate.append(this_class)
    elif EILphases[idx] == 'pre expert':
        class_preexpert.append(this_class)
    elif EILphases[idx] == 'post early':
        class_postearly.append(this_class)
    elif EILphases[idx] == 'post late':
        class_postlate.append(this_class)
    elif EILphases[idx] == 'post expert':
        class_postexpert.append(this_class)
```
