```python
#https://direct.mit.edu/neco/article/21/2/397/8692/Generating-Spike-Trains-with-Specified-Correlation

def phi(g):
    return 0.5*(1+erf(g/np.sqrt(2)))
def gamma(r):
    return np.sqrt(2) * erfinv(2*r - 1)

def phi_2(gamma_i,gamma_j,lambda_ij):
    mean = np.array([0,0])
    cov = np.array([[1, lambda_ij], [lambda_ij, 1]])
    X = np.array([gamma_i,gamma_j]).T
    return multivariate_normal.cdf(X,mean=mean,cov=cov)

def func(gamma_i,gamma_j,sigma_ij,lambda_ij):
    known_component = sigma_ij + phi(gamma_i)*phi(gamma_j)
    return known_component - phi_2(gamma_i,gamma_j,lambda_ij)
    
def solve_lambda_ij(gamma_i,gamma_j,sigma_ij):
    lower_bound = -0.9999#-1*phi(gamma_i)*phi(gamma_j)
    upper_bound = 0.9999#min(phi(gamma_i)*(1 - phi(gamma_j)), phi(gamma_j)*(1 - phi(gamma_i)))
    
    f = lambda l_ij: np.abs(func(gamma_i,gamma_j,sigma_ij,l_ij))
    return minimize_scalar(f,bounds=[lower_bound,upper_bound],method='bounded').x

def pseudo_flatten_rates(firing_rates,covariance,n_trials,verbose=False):
    n_clusts = len(firing_rates)
    if type(covariance) in [int,float,np.int64,np.float64]:
        covariance = np.full((n_clusts,n_clusts),covariance)
    assert covariance.shape[0] == covariance.shape[1]
    assert covariance.shape[0] == len(firing_rates)
    
    trial_length_sec=2.5
    fs=30000
    bin_length_msec=50
    pre_stim_padding_sec=0.2
    refractory_period_msec=2
    
    n_bins = int((2.5+pre_stim_padding_sec)*1000 / bin_length_msec)
    n_frames = int((2.5+pre_stim_padding_sec)*1000 / refractory_period_msec)
    
    clusts = list(range(n_clusters))
    trials = list(range(n_trials))
    ################################################################################################################
    
    #Calculate mean of distribution
    firing_rates_prob = [fr*refractory_period_msec/1000 for fr in firing_rates]
    gamma_vec = [gamma(frp) for frp in firing_rates_prob]
    
    #Calculate cov_mat of distribution
    lambda_mat = np.full_like(cov_mat,np.nan)
    #Start with diagonals
    lambda_mat = np.eye(n_clusts,n_clusts)
    #And now off-diagonals
    for i,j in product(range(n_clusts),range(n_clusts)):
        if i != j:
            if covariance[i,j] < 1e-10:
                lambda_mat[i,j] = 0
            else:
                lambda_mat[i,j] = covariance[i,j]#solve_lambda_ij(gamma_vec[i],gamma_vec[j],covariance[i,j])
    
    if verbose:
        for i in range(n_clusts):
            print(f"r_{i} is {phi(gamma_vec[i])}")
        for i in range(n_clusts):
            print(f"gamma_{i} is {gamma_vec[i]}")
        for i in range(n_clusts):
            print(f"variance {i} is {phi(gamma_vec[i])*phi(-gamma_vec[i])}")
        for i,j in product(range(n_clusts),range(n_clusts)):
            print(f"lambda_{i},{j} is {lambda_mat[i,j]}")
    
    ################################################################################################################
    
    number_of_points = int(n_trials*(trial_length_sec+pre_stim_padding_sec)*1000/refractory_period_msec)
    rates = np.random.multivariate_normal(gamma_vec,lambda_mat,size=number_of_points).T
    rates = np.greater(rates,0)
    
    #Binning
    ################################################################################################################
    
    frames_per_bin = int(bin_length_msec / refractory_period_msec)
    n_bins = int(np.ceil(n_trials*(trial_length_sec+pre_stim_padding_sec)*1000/bin_length_msec))
    FR_bins = np.full((n_clusts,n_bins),np.nan)
    for clust_idx in range(n_clusts):
        for bin_idx in range(n_bins):
            FR_bins[clust_idx,bin_idx] = np.sum(rates[clust_idx,bin_idx*frames_per_bin : (bin_idx*frames_per_bin + frames_per_bin)])
    
    ################################################################################################################

    return rates
```

```python
def calc_dim(N,v,cv):
    num = 1 + (N-1)*v
    den = 1 + (N-1)*cv
    return num/den

def calc_dim_GT(cov):
    return(np.trace(cov)**2 / np.trace(cov@cov))
```

```python
iterations = 100
N = range(2,100)

dim_var_res = dict()
for n in N:
    dim_var_res[n] = []
dim_covar_res = dict()
for n in N:
    dim_covar_res[n] = []
dim_full_res = dict()
for n in N:
    dim_full_res[n] = []
dim_GT_res = dict()
for n in N:
    dim_GT_res[n] = []

################################################################################################################################

for n_clust in tqdm(N):
    for iteration in range(iterations):
        firing_rates = np.random.exponential(1,size=n_clust)+0.1
        #firing_rates = np.array([1]*n_clust)
        rates = pseudo_flatten_rates(firing_rates,0.95,50,verbose=False)
        cov_mat = np.cov(rates)

        variances = cov_mat[np.eye(n_clust,n_clust,dtype='bool')]
        covariances = cov_mat[~np.eye(n_clust,n_clust,dtype='bool')]

        avg_sq_vars = np.mean(variances**2)
        avg_sq_covars = np.mean(covariances**2)
        avg_pred_covars = np.mean(np.outer(variances,variances)[~np.eye(n_clust,n_clust,dtype='bool')])

        dim_var_only = calc_dim(n_clust,avg_pred_covars/avg_sq_vars,0)
        dim_covar_only = calc_dim(n_clust,1,avg_sq_covars/avg_sq_vars)
        dim_full = calc_dim(n_clust,avg_pred_covars/avg_sq_vars,avg_sq_covars/avg_sq_vars)
        dim_GT = calc_dim_GT(cov_mat)

        dim_var_res[n_clust].append(dim_var_only/n_clust)
        dim_covar_res[n_clust].append(dim_covar_only/n_clust)
        dim_full_res[n_clust].append(dim_full/n_clust)
        dim_GT_res[n_clust].append(dim_GT/n_clust)
```

```python
fig = plt.figure(figsize=(5,5))
gs = fig.add_gridspec(2,2,hspace=0.5)
axs = [plt.subplot(gs[i,j]) for i,j in product(range(2),range(2))]

dims = [dim_var_res,dim_covar_res,dim_full_res,dim_GT_res]
for ax,dim in zip(axs,dims):
    Xs = N
    Ys = np.array([np.nanmean(dim[d]) for d in Xs])
    Y_sem = np.array([sem(ilep.rmnan(dim[d])) for d in Xs])
    
    ax.plot(Xs,Ys,color='k')
    ax.fill_between(Xs,Ys-Y_sem,Ys+Y_sem,color='k',alpha=0.25,lw=0)
    
    ax.set_xlim([0,100])
    ax.set_ylim([0,1])
```
