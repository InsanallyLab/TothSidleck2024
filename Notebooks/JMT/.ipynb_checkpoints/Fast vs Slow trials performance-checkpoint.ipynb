```python
EnumSession = []
EnumClust = []
sessions = os.listdir(act_directory)
for session in tqdm(sessions):
    sessionfile = ilep.loadSessionCached(act_directory,session)
    
    if sessionfile.meta.task in ['passive no beahvior']:
        continue
    if sessionfile.meta.task in ['tuning nonreversal','tuning switch','tuning reversal']:
        continue
    if sessionfile.meta.region != 'AC':
        continue
        
    if sessionfile.meta.animal not in animals:
        continue
    
    #for clust in sessionfile.clusters.good:
    EnumSession.append(session)
    #EnumClust.append(clust)
```

```python
slow_pcs = []
fast_pcs = []

for session in tqdm(EnumSession):
    sessionfile = ilep.loadSessionCached(act_directory,session)
    all_conditions = ilep.getAllConditions(sessionfile,clust=None)
    
    # Slow_go
    slow_go_trials = all_conditions['slow_go'].trials
    hit_trials = all_conditions['hit'].trials
    slow_pc = np.sum(np.isin(slow_go_trials,hit_trials)) / len(slow_go_trials)
    
    # Slow_go
    fast_go_trials = all_conditions['fast_go'].trials
    hit_trials = all_conditions['hit'].trials
    fast_pc = np.sum(np.isin(fast_go_trials,hit_trials)) / len(fast_go_trials)
    
    slow_pcs.append(slow_pc)
    fast_pcs.append(fast_pc)
    
slow_pcs = np.array(slow_pcs)*100
fast_pcs = np.array(fast_pcs)*100
```

```python
fig, ax = plt.subplots(figsize=(4,4.5))

X = np.random.uniform(-0.25,0.25,size=len(slow_pcs))
ax.scatter(X,slow_pcs,c='purple',s=10)#,marker='s')
ax.scatter(X+1,fast_pcs,c='green',s=10)#,marker='s')
ax.plot([X,X+1],[slow_pcs,fast_pcs],c=GOOD_GREY,linestyle='-',lw=0.5,zorder=-10)

ax.set_xlim([-0.5,1.5])
ax.set_xticks([0,1])
ax.set_xticklabels(['Slow','Fast'])

ax.set_ylim([0,100])
ax.set_ylabel('Percent correct (%)')

ax.scatter([],[], ec='purple',fc='purple', s=10,lw=1, zorder=10,label='Slow responses')# per session',marker='s')
ax.scatter([],[], ec='green',fc='green', s=10,lw=1, zorder=10,label='Fast responses')# per session',marker='s')
#lgd = ax.legend(frameon=False,loc=(0.6,0.05))
lgd = ax.legend(frameon=False,loc=(1,0.8))

fig.savefig(os.path.join('D:\\\\Figures','Manuscript Figures','Performance on fast vs slow trials V'+str(current_version)+'.pdf'),transparent=False,facecolor="white", bbox_extra_artists=(lgd,), bbox_inches='tight')
```
