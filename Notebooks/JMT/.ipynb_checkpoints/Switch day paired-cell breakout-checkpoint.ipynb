```python
def stat_test_increased_decreased_baseline(sessionfile,clust):
    all_conditions = ilep.getAllConditions(sessionfile,clust,trialsPerDayLoaded='NO_TRIM')
    
    pre_trials = all_conditions['pre_switch'].trials
    post_trials = all_conditions['post_switch'].trials
    
    pre_spikes = []
    for trial in pre_trials:
        starttime = sessionfile.trials.starts[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.starts[trial]
        trial_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        pre_spikes.append(trial_spikes)
    
    post_spikes = []
    for trial in post_trials:
        starttime = sessionfile.trials.starts[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.starts[trial]
        trial_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        post_spikes.append(trial_spikes)
        
    try:
        pval = mannwhitneyu(pre_spikes,post_spikes).pvalue
        if pval > 0.05:
            return 0,pval
        elif pval <= 0.05 and np.mean(post_spikes)>np.mean(pre_spikes):
            return 1,pval
        elif pval <= 0.05 and np.mean(post_spikes)<np.mean(pre_spikes):
            return -1,pval
    except Exception as e:
        print(e)
        return np.nan,np.nan
```

```python
def stat_test_increased_decreased_FR(sessionfile,clust):
    all_conditions = ilep.getAllConditions(sessionfile,clust,trialsPerDayLoaded='NO_TRIM')
    
    try:
        peakstart_pre = sessionfile.responsiveness[clust]['pre_switch'].peakstart
        peakend_pre = sessionfile.responsiveness[clust]['pre_switch'].peakend
        peakstart_post = sessionfile.responsiveness[clust]['post_switch'].peakstart
        peakend_post = sessionfile.responsiveness[clust]['post_switch'].peakend
    except:
        return np.nan,np.nan
        
    pre_trials = all_conditions['pre_switch'].trials
    post_trials = all_conditions['post_switch'].trials
    
    pre_spikes = []
    for trial in pre_trials:
        starttime = sessionfile.trials.starts[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.starts[trial]
        baseline_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        baseline_FR = baseline_spikes / (100)
        
        starttime = sessionfile.trials.starts[trial] + (peakstart_pre*30)#Convert ms to samples
        endtime = sessionfile.trials.starts[trial] + (peakend_pre*30)#Convert ms to samples
        trial_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        trial_FR = trial_spikes / (peakend_pre-peakstart_pre)
        
        pre_spikes.append(trial_FR-baseline_FR)
    
    post_spikes = []
    for trial in post_trials:
        starttime = sessionfile.trials.starts[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.starts[trial]
        baseline_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        baseline_FR = baseline_spikes / (100)
        
        starttime = sessionfile.trials.starts[trial] + (peakstart_post*30)#Convert ms to samples
        endtime = sessionfile.trials.starts[trial] + (peakend_post*30)#Convert ms to samples
        trial_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        trial_FR = trial_spikes / (peakend_post-peakstart_post)
        
        post_spikes.append(trial_FR-baseline_FR)
        
    try:
        pval = mannwhitneyu(pre_spikes,post_spikes).pvalue
        if pval > 0.05:
            return 0,pval
        elif pval <= 0.05 and np.mean(post_spikes)>np.mean(pre_spikes):
            return 1,pval
        elif pval <= 0.05 and np.mean(post_spikes)<np.mean(pre_spikes):
            return -1,pval
    except Exception as e:
        print(e)
        return np.nan,np.nan
```

```python
def stat_test_increased_decreased_ramping(sessionfile,clust):
    all_conditions = ilep.getAllConditions(sessionfile,clust,trialsPerDayLoaded='NO_TRIM')
    
    pre_trials = all_conditions['pre_switch_go'].trials
    post_trials = all_conditions['post_switch_go'].trials
    
    pre_spikes = []
    for trial in pre_trials:
        starttime = sessionfile.trials.starts[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.starts[trial]
        baseline_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        starttime = sessionfile.trials.response[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.response[trial]
        trial_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        pre_spikes.append(trial_spikes-baseline_spikes)
    
    post_spikes = []
    for trial in post_trials:
        starttime = sessionfile.trials.starts[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.starts[trial]
        baseline_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        starttime = sessionfile.trials.response[trial] - 100 * sessionfile.meta.fs / 1000
        endtime = sessionfile.trials.response[trial]
        trial_spikes = len(ilep.getSpikeTimes(sessionfile,clust=clust,starttime=starttime,endtime=endtime))
        post_spikes.append(trial_spikes-baseline_spikes)
        
    try:
        pval = mannwhitneyu(pre_spikes,post_spikes).pvalue
        if pval > 0.05:
            return 0,pval
        elif pval <= 0.05 and np.mean(post_spikes)>np.mean(pre_spikes):
            return 1,pval
        elif pval <= 0.05 and np.mean(post_spikes)<np.mean(pre_spikes):
            return -1,pval
    except Exception as e:
        print(e)
        return np.nan,np.nan
```

```python
def many_paired_rmnan(Xs,positiveOnly=False):
    '''
    removes nans from paired data. Requires 1-dimension. Returns as ndarrays.
    '''
    mask = np.full_like(Xs[0],True)
    
    for X in Xs:
        X = np.array(X)
        if len(X.shape) != 1:
            raise Exception(f"X is of dimension {len(X.shape)} but only 1-dimensional data is supported")
        mask = np.logical_and(np.isfinite(X),mask)
        if positiveOnly:
            mask = np.logical_and(mask,np.greater(X,0))

    ret = [np.array(X)[mask] for X in Xs]
    return ret
```
