```python
for idx,(session,clust) in tqdm(enumerate(zip(EnumACSession,EnumACClust))):
    ACsessionfile = ilep.loadSessionCached(act_directory,session)
    
    #1,2,3,4
    if ACsessionfile.meta.day_of_training <= pre_early:
        passive_FRmod_preearly.append(ACsessionfile.responsiveness[clust]['all_trials'].FRmodulation)
    #5,6
    if ACsessionfile.meta.day_of_training <= pre_late and ACsessionfile.meta.day_of_training >pre_early:
        passive_FRmod_prelate.append(ACsessionfile.responsiveness[clust]['all_trials'].FRmodulation)
    #7,8
    if ACsessionfile.meta.day_of_training <= pre_expert and ACsessionfile.meta.day_of_training >pre_late:
        passive_FRmod_preexpert.append(ACsessionfile.responsiveness[clust]['all_trials'].FRmodulation)
    #9,10,11,12
    if ACsessionfile.meta.day_of_training <= post_early and ACsessionfile.meta.day_of_training >pre_expert:
        passive_FRmod_postearly.append(ACsessionfile.responsiveness[clust]['all_trials'].FRmodulation)
    #13,14,15
    if ACsessionfile.meta.day_of_training <= post_late and ACsessionfile.meta.day_of_training >post_early:
        passive_FRmod_postlate.append(ACsessionfile.responsiveness[clust]['all_trials'].FRmodulation)
    #16,17,18,19,20,21
    if ACsessionfile.meta.day_of_training >= post_expert:
        passive_FRmod_postexpert.append(ACsessionfile.responsiveness[clust]['all_trials'].FRmodulation)
```

```python
data = [[passive_FRmod_preearly,passive_FRmod_prelate,passive_FRmod_preexpert],[passive_FRmod_postearly,passive_FRmod_postlate,passive_FRmod_postexpert]]
stage = ['prereversal','postreversal']
labels = [['Early','Late','Expert'],['Early','Late','Expert']]
x_range = np.linspace(0,100,1000)


for stage_idx,(ax,col_data) in enumerate(zip(axs,data)):
    fig = plt.figure(figsize=(4,4.5))
    bax = brokenaxes(xlims=( (0,20), (20, 100) ), width_ratios=[3,1], wspace=0.025)
    
    ymax=-np.Inf
    for idx,col in enumerate(col_data):
        KDE = gaussian_kde(ilep.rmnan(col), bw_method='scott')
        KDE_eval = KDE.evaluate(x_range)
        #KDE_eval = np.cumsum(KDE_eval) / np.sum(KDE_eval)
        bax.plot(x_range,KDE_eval,color=PHASE_COLORS[idx+3*stage_idx],label=labels[stage_idx][idx],lw=3)
        ymax=max(ymax,np.max(KDE_eval))
    
    bax.set_xlabel('Firing rate modulation (spikes/s)')
    bax.axs[0].set_ylabel('Probability density (normalized)')
    
    bax.set_ylim([0,ymax*1.01])
    bax.axs[0].set_yticks(bax.axs[0].get_ylim())
    bax.axs[0].set_yticklabels([0,1])

    bax.legend(frameon=False,loc='center right')

    bax.axs[0].set_xticks([0,10,20])
    bax.axs[1].set_xticks([40,60,80,100])
    
    fig.savefig(os.path.join('D:\\\\Figures','Manuscript Figures','Passive FRmodulation distribution broken axes '+stage[stage_idx]+' V'+str(current_version)+'.pdf'),transparent=False,facecolor="white")
```

### V4

```python
data = [[passive_FRmod_preearly,passive_FRmod_prelate,passive_FRmod_preexpert],[passive_FRmod_postearly,passive_FRmod_postlate,passive_FRmod_postexpert]]
stage = ['prereversal','postreversal']
labels = [[1,2,3],[4,5,6]]
x_range = np.linspace(0,100,1000)

colors = [GOOD_GREY,'grey','k',GOOD_GREY,'grey','k']

for stage_idx,(col_data) in enumerate(data):
    fig = plt.figure(figsize=(2.5,4.5/2))
    bax = brokenaxes(xlims=( (0,20), (20, 100) ), width_ratios=[3,1], wspace=0.025)
    
    ymax=-np.Inf
    for idx,col in enumerate(col_data):
        KDE = gaussian_kde(ilep.rmnan(col), bw_method='scott')
        KDE_eval = KDE.evaluate(x_range)
        #KDE_eval = np.cumsum(KDE_eval) / np.sum(KDE_eval)
        bax.plot(x_range,KDE_eval,color=colors[idx+3*stage_idx],label=labels[stage_idx][idx],lw=3)
        ymax=max(ymax,np.max(KDE_eval))
    
    xlab = bax.set_xlabel('Firing rate modulation (spikes/s)',fontsize=14)
    ylab = bax.axs[0].set_ylabel('Probability density',fontsize=14)
    
    bax.set_ylim([0,ymax*1.01])
    bax.axs[0].set_yticks(bax.axs[0].get_ylim())
    bax.axs[0].set_yticklabels([0,1])

    bax.legend(frameon=False,loc='center right')

    bax.axs[0].set_xticks([0,10,20])
    bax.axs[1].set_xticks([60,100])
    
    for item in ([bax.axs[0].title, bax.axs[0].xaxis.label, bax.axs[0].yaxis.label] +
        bax.axs[0].get_xticklabels() + bax.axs[0].get_yticklabels()):
        item.set_fontsize(stdfont)
    for item in ([bax.axs[1].title, bax.axs[1].xaxis.label, bax.axs[1].yaxis.label] +
        bax.axs[1].get_xticklabels() + bax.axs[1].get_yticklabels()):
        item.set_fontsize(stdfont)
    
    artists=[xlab,ylab]
    fig.savefig(os.path.join('D:\\\\Figures','Manuscript Figures','Passive FRmodulation distribution broken axes '+stage[stage_idx]+' V'+str(current_version)+'.pdf'),transparent=False,facecolor="white",dpi=400, bbox_extra_artists=(artists), bbox_inches='tight')
```
