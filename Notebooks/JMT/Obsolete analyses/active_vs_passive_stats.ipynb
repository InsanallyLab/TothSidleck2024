```python
import pandas as pd
import pingouin as pg
```

```python
df = pd.read_csv('C:\\\\Users\\insan\\Desktop\\DataForStatsTests.csv', index_col=0)
```

```python
df.columns
```

```python
df.phase.unique()
```

```python
pg.anova(data=df, dv='FR_mod', between=['phase', 'context'], detailed=True)
```

```python
context_results = pg.pairwise_tests(data=df, dv='FR_mod', between=['phase', 'context'], parametric=True, padjust='fdr_bh')
context_results
```

```python
context_results[context_results["p-corr"] < 0.05]
```

```python
phase_results = pg.pairwise_tests(data=df, dv='FR_mod', between=['context', 'phase'], parametric=False, padjust='fdr_bh')
```

```python
phase_results[phase_results["p-corr"] < 0.05]
```
