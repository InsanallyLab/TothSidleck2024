```python
directory = 'S:\\Sorts\DualRecording'
animals = os.listdir(directory)
for animal in tqdm(animals):
    sessions = os.listdir(os.path.join(directory,animal))
    for session in sessions:
        for region in ['AC','M2','MGB','Striatum']:
            model_filename = os.path.join(directory,animal,session,region,'params.py')
            clusters_filename = os.path.join(directory,animal,session,region,'cluster_info.tsv')
            dat_filename = os.path.join(directory,animal,session,region,'temp_wh.dat')
            if os.path.isfile(model_filename):
                print(f"animal {animal} session {session} region {region}")
                
                model = load_model(model_filename)
                model.dat_path = [Path(dat_filename)]
                model._load_data()
                
                cluster_info = pd.read_csv(clusters_filename,delimiter='\t')
                clusters = list(cluster_info['id'])
                
                waveforms = dict()
                for clust in tqdm(clusters):
                    clust_waveforms = model.get_cluster_spike_waveforms(clust)
                    waveforms[clust] = clust_waveforms
                    
                output_filename = os.path.join(directory,animal,session,region,'waveforms.pickle')
                with open(output_filename, 'wb') as f:
                    pickle.dump(waveforms, f, protocol=pickle.HIGHEST_PROTOCOL)
```
